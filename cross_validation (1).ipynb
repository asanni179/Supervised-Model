```python
import pandas as pd
```

```python
adult = pd.read_csv('/content/drive/MyDrive/adult.csv')
```

```python
adult.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
adult['income'] = lb.fit_transform(adult['income'])
adult['relationship'] = lb.fit_transform(adult['relationship'])
adult['marital-status'] = lb.fit_transform(adult['marital-status'])
adult['education'] = lb.fit_transform(adult['education'])
adult['workclass'] = lb.fit_transform(adult['workclass'])
adult['occupation'] = lb.fit_transform(adult['occupation'])
```

```python
adult = pd.get_dummies(adult, columns = ['race', 'gender', 'native-country'])
```

```python
adult.head()
```

```python
adult.info()
```

```python
y = adult['income']
y
```

```python
x = adult.drop('income', axis = 1)
x
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

cv_model_rf= RandomForestClassifier(n_estimators = 198, max_depth= None)
cv_model_dt = DecisionTreeClassifier()
cv_model_lg = LogisticRegression()

scores_rf = cross_val_score(cv_model_rf,x,y,cv = 4)
scores_dt = cross_val_score(cv_model_dt,x,y,cv = 4)
scores_lg = cross_val_score(cv_model_lg,x,y,cv = 4)

print(f'The final accuracy for the random forest model is {scores_rf.mean()}')
print(f'The final accuracy for the Decision tree model is {scores_dt.mean()}')
print(f'The final accuracy for the linear regression model is {scores_lg.mean()}')
```
